```python
import pandas as pd
import matplotlib as plt

%matplotlib inline
%config InlineBackend.figure_format ='retina'

murders = pd.read_csv("Homicide_Statistics.csv")
murders
```

```python
murders = murders.drop(columns = 'Unnamed: 0')
murders
```

```python
plot1 = murders[['City','Firearm Homicides','Non-Firearm Homicides','Unspecified Weapon Homicides']].plot.bar(stacked=True, legend=True,figsize=(5,10), colormap="nipy_spectral", xlabel = 'City', ylabel = 'Total Homicides', title = 'City Homicides by Weapon')                                                                  
plot1.set_xticklabels(murders['City']);

#<AxesSubplot:title={'center':'City Homicides by Weapon'}, xlabel='City', ylabel='Total Homicides'>
```

```python
plot2 = murders[['City','Firearm Homicide Arrests','Non-Firearm Homicide Arrests','Unspecified Weapon Homicide Arrests']].plot.bar(stacked=True, legend=True,figsize=(10,18), colormap="nipy_spectral", xlabel = 'City', ylabel = 'Total Arrests', title = 'City Homicide Arrests by Weapon')                                                                  
plot2.set_xticklabels(murders['City']);
    
```

```python
plot3 = murders[['City','Firearm Homicide Arrest Rate(%)','Non-Firearm Homicide Arrest Rate(%)']].plot.bar(stacked=False, legend=True,figsize=(8,8), xlabel = 'City', ylabel = 'Arrest Rate(%)', title = 'City Homicide Arrest Rates by Weapon', color = ["red","blue"]) 
plot3.set_xticklabels(murders['City']);
```

This bar graph comapres the arrest rate of homicides involving guns to those commited with other weapons in 11 different cities

I went through multiple iterations when creating my plots while trying to figure out how to best visualize the data. I ultimately had to go back and create new colums in my dataset that calculated arrest rate. This allowed me to plot and compare multiple arrest rates from each city. This makes it very easy to see that cases involving firearms are solved at a lowder rate in all the cities with sufficient data.
